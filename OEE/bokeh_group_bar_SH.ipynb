使用bokeh绘制各制造产地每周的护套（SH）产值比例的多组柱状图

```python
from bokeh.io import show, output_notebook
output_notebook()
from bokeh.models import ColumnDataSource, FactorRange
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.core.properties import value
```

```python
palette = ["#c9d9d3", "#718dbf", "#e84d60","#E6B8B7"]
week_number = df_SH['week']
pro_area = ['W','L','S','Y']
TOOLS="pan,wheel_zoom,box_zoom,reset,tap,hover"
data = {
    'week_number':week_number,
    'wuhan_sh':df_SH['wuhan_sh'],
    'lanzhou_sh':df_SH['lanzhou_sh'],
    'shenyang_sh':df_SH['shenyang_sh'],
    'yinni_sh':df_SH['yinni_sh'],
}

area_list = []
a = 1
while a <= 50:
    area_list.append('武汉')
    area_list.append('兰州')
    area_list.append('沈阳')
    area_list.append('印尼')
    a+=1
    
# print(area_list)
x = [(number,area)for number in week_number for area in pro_area]
# x = [(number,area)for number in week_number for area in pro_area]
counts = sum(zip(data['wuhan_sh'],data['lanzhou_sh'],data['shenyang_sh'],data['yinni_sh']),())

source = ColumnDataSource(data=dict(x=x, counts=counts,label = area_list))
p = figure(x_range=FactorRange(*x), plot_height=350,plot_width=2500, 
           title="各制造产地每周的护套（SH）产值比例的多组柱状图(W:武汉、L:兰州、S:沈阳、Y:印尼)",
           toolbar_location=None, tools=TOOLS)
p.hover.tooltips = [('周数/制造商',"@x"),('产值比例',"@counts")]
p.vbar(x='x', top='counts', width=1, source=source, line_color="white",
      fill_color=factor_cmap('x', palette=palette, factors=pro_area, start=1, end=2),legend = 'label')
p.y_range.start = 0
p.x_range.range_padding = 0.005
p.xaxis.major_label_orientation = 0.5
p.xgrid.grid_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"

show(p)
```
